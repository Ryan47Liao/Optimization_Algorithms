Least Square for cubic function:

Example 1:

We first generate 100 random sample points on the plane by $y=x^3-2x^2-5x+6+e$ 
where $e\sim N(0,1)$. Then we use linear least square to find the best estimation for the coefficients.

Direct Method:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# X ranges from [0, 4]
X = 4 * np.random.rand(100,1)
y = X**3 - 2*X**2 - 5*X + 6 + np.random.randn(100,1)
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=14)
plt.axis([0, 4, -10, 20])
xx = np.linspace(0,4,100) # for plotting
yy = xx**3 - 2*xx**2 - 5*xx + 6 # for plotting
plt.plot(xx, yy, "g--")
plt.show()
```

In statistics, the coefficient of determination, denoted $R^2$ or $r^2$ and pronounced "R squared", is the proportion of the variance in the dependent variable that is predictable from the independent variable. It is used to analyze how differences in one variable can be explained by a difference in a second variable. $R^2 = \frac{\big[\sum{(y_i-\bar{y})(x_i-\bar{x})}\big]^2}{\sum{(y_i-\bar{y})^2}\sum{(x_i-\bar{x})^2}}$, here $\bar{x}=\sum{x_i}/n$, and we have $0\leq R^2 \leq 1$. 

Now compute the $R^2$ for the following 2 samples. Conclude what you find.

```python
# 1
# y = x + e, e obeys 0.2*N(0,1)
X1 = 4 * np.random.rand(50,1) - 2 # X ranging from -2 to 2
y1 = X1  + 0.2*np.random.randn(50,1)

plt.plot(X1, y1, "b.")
plt.xlabel("$x_1$", fontsize=14)
plt.ylabel("$y$", rotation=0, fontsize=14)
plt.axis([-3, 3, -3, 3])
xx = np.linspace(-2,2,50) # for plotting
yy = xx # for plotting
plt.plot(xx, yy, "g--")
plt.show()
```
